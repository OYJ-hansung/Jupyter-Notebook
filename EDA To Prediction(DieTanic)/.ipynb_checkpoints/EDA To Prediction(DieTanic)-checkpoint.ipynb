# Feature Analysis

## Categorical Features
범주형, 모든 value를 정해진 변수로 나눌 수 있는 Feature
ex) Sex, Embarked

## Ordinal Features
순서형, 범주의 순서에 의미를 둘 수 있는 Feature
ex)Pclass(1, 2, 3), size(Tall, Medium, Short)

## Continuous Features
연속형, 연속적인 값
ex)Age

# 1. Sex (Categorical Feature)

```python
data.groupby(['Sex','Survived'])['Survived'].count()

# Sex + Survived를 그룹으로 묶고, Survived의 누적값을 확인한 결과
# 여자 생존 - 233 / 사망 - 81
# 남자 생존 - 109 / 사망 - 468
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot('Sex',hue='Survived',data=data,ax=ax[1])
ax[1].set_title('Sex:Survived vs Dead')
plt.show()

# 다시 한번 figure를 정의하고 위에 있는 데이터를 막대 그래프로 시각화.
# 여성의 생존률이 월등히 높았다.
```

# 2. Pclass (Ordinal Feature)

```python
pd.crosstab(data.Pclass,data.Survived,margins=True).style.background_gradient(cmap='summer_r')
# pd.crosstab(data.Pclass,data.Survived,margins=False).style.background_gradient(cmap='summer_r')

# 객실 클래스 3(가장 높은), 2, 1(가장 낮은) 순서
# crosstab으로 교차표 생성 및 value_count 후 background_gradient로 heatmap 작성
# 생존자가 가장 많은 class는 1(136), 사망자가 가장 많은 클래스는 3(372)
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Number Of Passengers By Pclass')
ax[0].set_ylabel('Count')
sns.countplot('Pclass',hue='Survived',data=data,ax=ax[1])
ax[1].set_title('Pclass:Survived vs Dead')
plt.show()

# Pclass value_count로 승객의 수를 비교하기 위한 첫 axis
# 각 클래스마다 생존자의 비율을 확인하기 위한 두번째 axis
```

```python
data[['Pclass', 'Survived']].groupby('Pclass').mean()

# 두번째 axis의 생존률을 수치화하면 다음과 같다. 생존률 부자 > 가난
```

```python
pd.crosstab([data.Sex,data.Survived],data.Pclass,margins=True).style.background_gradient(cmap='summer_r')

# 히트맵으로 이번엔 각 Pclass별 sex의 생존률을 확인
```

```python
sns.factorplot('Pclass','Survived',hue='Sex',data=data)
plt.show()

# factorplot는 categorical feature을 시각화하는데 자주 사용되는 그래프입니다
# 히트맵과 factorplot를 확인한 결과 Pclass와 무관하게 남성의 사망률이 월등히 높다.
```

# 3. Age (Continuous Feature)

```python
print('Oldest Passenger was of:',data['Age'].max(),'Years')
print('Youngest Passenger was of:',data['Age'].min(),'Years')
print('Average Age on the ship:',data['Age'].mean(),'Years')

# 승객들중 나이가 가장 많은 사람, 가장 어린 사람, 평균 나이
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
sns.violinplot("Pclass","Age", hue="Survived", data=data,split=True,ax=ax[0])
ax[0].set_title('Pclass and Age vs Survived')
ax[0].set_yticks(range(0,110,10))
sns.violinplot("Sex","Age", hue="Survived", data=data,split=True,ax=ax[1])
ax[1].set_title('Sex and Age vs Survived')
ax[1].set_yticks(range(0,110,10))
plt.show()

# Pclass와 Age에 따른 생존률 시각화 by violin plot - box plot과 유사
# 어린이들(0 - 10세)와 노인들(65세 이상)의 생존률을 보면 어린이들의 생존률은 높았고 노인의 생존률은 낮은 모양
# 
```

```python
data['Initial']=0
for i in data:
    data['Initial']=data.Name.str.extract('([A-Za-z]+)\.')

# 처음에 살펴봤을 때, Age에 Null값이 존재. 이것을 채우기 위한 작업
# extract로 A-Z와 a-z의 문자중 다음 문자로 .이 따라오는 것을 추출해서 새로운 Feaure Initial 칼럼 생성
```

```python
pd.crosstab(data.Initial,data.Sex).T.style.background_gradient(cmap='summer_r')

# 히트맵을 통해서 Inital을 확인해보면, 데이터 중에서 철자가 잘못된 이름들이 있다. Mlle or Mme that stand for Miss. 
```

```python
data['Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],inplace=True)

# Initial을 수정.
```

```python
data.groupby('Initial')['Age'].mean() #lets check the average age by Initials

# 각 Initial마다의 나이 평균을 구해서 Age의 Null값을 채운다.
```

```python
data.loc[(data.Age.isnull())&(data.Initial=='Mr'),'Age']=33
data.loc[(data.Age.isnull())&(data.Initial=='Mrs'),'Age']=36
data.loc[(data.Age.isnull())&(data.Initial=='Master'),'Age']=5
data.loc[(data.Age.isnull())&(data.Initial=='Miss'),'Age']=22
data.loc[(data.Age.isnull())&(data.Initial=='Other'),'Age']=46

data.Age.isnull().any()
# 더 이상 Age에는 null값이 없다.
```

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
data[data['Survived']==0].Age.plot.hist(ax=ax[0],bins=20,edgecolor='black',color='red')
ax[0].set_title('Survived= 0')
x1=list(range(0,85,5))
ax[0].set_xticks(x1)
data[data['Survived']==1].Age.plot.hist(ax=ax[1],color='green',bins=20,edgecolor='black')
ax[1].set_title('Survived= 1')
x2=list(range(0,85,5))
ax[1].set_xticks(x2)
plt.show()

# Null값도 채웠기 때문에, Age와 사망률에 대한 연관성을 살펴본다.
# 아이들의 생존이 우선시 되었다. 가장 노인들 (80세)의 생존이 우선시 되었다.
```

```python
sns.factorplot('Pclass','Survived',col='Initial',data=data)
plt.show()

# Initial에 따른 Pclass 생존율을 factorplot으로 확인
# 남자를 나타내는 Mr의 생존률이 극히 낮았다. 여성을 나타내는 Initial의 생존율이 가장 높다.
```

# 4. Embarked (Categorical Feature)

```python
pd.crosstab([data.Embarked,data.Pclass],[data.Sex,data.Survived],margins=True).style.background_gradient(cmap='summer_r')

# 승객이 승선을 한 항구에 따른 생존율 확인
# S에서 승선한 남자들의 사망자 인원이 눈에 띄게 많았다.
```

```python
sns.factorplot('Embarked','Survived',data=data)
fig=plt.gcf()
fig.set_size_inches(5,3)
plt.show()

# S의 생존률이 가장 낮다. C의 생존률이 가장 높다.
```

```python
f,ax=plt.subplots(2,2,figsize=(20,15))
sns.countplot('Embarked',data=data,ax=ax[0,0])
ax[0,0].set_title('No. Of Passengers Boarded')
sns.countplot('Embarked',hue='Sex',data=data,ax=ax[0,1])
ax[0,1].set_title('Male-Female Split for Embarked')
sns.countplot('Embarked',hue='Survived',data=data,ax=ax[1,0])
ax[1,0].set_title('Embarked vs Survived')
sns.countplot('Embarked',hue='Pclass',data=data,ax=ax[1,1])
ax[1,1].set_title('Embarked vs Pclass')
plt.subplots_adjust(wspace=0.2,hspace=0.5)
plt.show()

# 1.1승선위치에 따른 승객수
# 1.2승선위치에 따른 성별
# 1.3승선위치에 따른 생존률
# 1.4승선위치에 따른 Pclass

# 아래 그래프를 통해서 s의 사망수가 많은 이유는 pclass 3의 사람들이 몰려 있었기
# 때문이라고 해도 될 것 같다.
```

```python
sns.factorplot('Pclass','Survived',hue='Sex',col='Embarked',data=data)
plt.show()
```
